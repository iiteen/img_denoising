```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Set the path to your folder
folder_path = '/content/drive/MyDrive/img_code/result'  # Replace with your folder path

# Create lists to store grouped file paths
gt_files = []
ori_files = []
out_files = []

# Iterate over all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith("_gt.png"):
        gt_files.append(os.path.join(folder_path, filename))
    elif filename.endswith("_ori.png"):
        ori_files.append(os.path.join(folder_path, filename))
    elif filename.endswith("_out.png"):
        out_files.append(os.path.join(folder_path, filename))

# Sort the lists to match files with the same sequence
gt_files.sort()
ori_files.sort()
out_files.sort()

# Function to display images side by side for comparison
def display_images(images):
    fig, axes = plt.subplots(1, len(images), figsize=(15, 5))
    for ax, img_path in zip(axes, images):
        img = Image.open(img_path)
        ax.imshow(img)
        ax.axis('off')
    plt.show()

# Display each group of images
for i in range(len(gt_files)):
    print(f"Comparing sequence: {gt_files[i]}")
    display_images([gt_files[i], ori_files[i], out_files[i]])
```
